```python
matrix = np.array([["even",7,1,5,3,6],
                  ["odd",4,6,2,5,7],
                  ["even",1,7,0,6,9],
                  ["odd",3,6,9,4,6],
                  ["even",1,0,8,5,8],
                  ["odd",1,5,8,2,5]])
df = pd.DataFrame(matrix, columns=["col0","col1","col2","col3","col4","col5"], index=["row1","row2","row3","row4","row5","row6"])
display(df.groupby(["col0"]).std())
```

##Завдання 3. 

3.1. Створимо лінійну залежність.

```python
from google.colab import files
import math
array = np.random.randn(100,5)
np.random.seed(5)
df2 = pd.DataFrame(array, columns=["x","y","y_hat","mAE","mSE"])
df2["y"] = df2["x"].apply(lambda x: x*2+1+(np.random.random()*2-1)/5)
df2["y_hat"] = df2["x"].apply(lambda x: x*2+1)
df2["mAE"] = df2["y"]-df2["y_hat"]
df2["mSE"] = (df2["y"]-df2["y_hat"])**2
display(df2)
df2.to_csv('filename.csv') 
files.download('filename.csv')

print("MAE:")
print(abs(df2["mAE"].sum()))
print("MSE:")
print(math.sqrt(df2["mSE"].sum()))
```

3.2. Побудуємо графік.

```python
plt.figure(figsize=(15,10)) 

x = df2["x"] 
y = df2["y"]
 
plt.scatter(x,y, color="blue") 

plt.plot([-3,2.5],[-5,6], color="red")


plt.title("plot") 
plt.xlabel("x") 
plt.ylabel("y")
```
